```python
#****************************************#
# 1. Lowercasing all text in the DataFrame
#****************************************#

# Lowercasing is a text preprocessing step where all letters in the text are converted to lowercase.
# This ensures the algorithm does not treat the same words differently due to casing.
df_copy = df.copy()

for col in df_copy.columns:
    df_copy[col] = [
        x.lower() if isinstance(x, str) else x
        for x in df_copy[col]
    ]


#****************************************#
# 2. Remove Punctuation & Special Characters
#****************************************#

import re


for col in df_copy.columns:
    df_copy[col] = [
        re.sub(r'[^a-zA-Z0-9\s]', ' ', x) if isinstance(x, str) else x
        for x in df_copy[col]
    ]


#****************************************#
# 3. Remove Stopwords
#****************************************#

import nltk
from nltk.corpus import stopwords

# Download stopwords if not already available
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))


for col in df_copy.columns:
    df_copy[col] = [
        ' '.join([word for word in x.split() if word not in stop_words])
        if isinstance(x, str) else x
        for x in df_copy[col]
    ]


#****************************************#
# 4. Full Text Cleaning
#****************************************#

# Apply your custom cleaning function to the 'file_content'.
from Functions import clean_text_column

df_copy['file_content'] = clean_text_column(df_copy, 'file_content')
```

```python
def lemmetization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
    nlp.max_length = 2_000_000
    texts_out = []
    for text in texts:
        doc = nlp(text)
        new_text =[]
        for token in doc:
            if token.pos_ in allowed_postags:
                new_text.append(token.lemma_)
        final = ' '.join(new_text)
        texts_out.append(final)
    return texts_out

lemmatized_texts = lemmetization(data)
print(lemmatized_texts[:1][0:90])
```

```python
# --- Imports ---
from gensim.models import Phrases, LdaModel, TfidfModel
from gensim.models.phrases import Phraser
from gensim import corpora
import pyLDAvis.gensim

# --- 1. Build Bigrams and Trigrams ---
# data_words: list of tokenized documents

# Bigram
bigram_phrases = Phrases(data_words, min_count=25, threshold=50)

bigram = Phraser(bigram_phrases)

def make_bigrams(texts):
    return [bigram[doc] for doc in texts]


# Apply transformations
data_bigrams = make_bigrams(data_words)
#data_bigrams_trigrams = make_trigrams(data_words)
data_bigrams = make_bigrams(data_words)


# --- 2. Build new Dictionary and Corpus ---
id2word = corpora.Dictionary(data_bigrams)
corpus = [id2word.doc2bow(text) for text in data_bigrams]

# --- 3. TF-IDF Filtering ---
tfidf = TfidfModel(corpus, id2word=id2word)
low_value = 0.03

filtered_texts = []

for bow in corpus:
    tokens = [id2word[id] for id, value in tfidf[bow] if value >= low_value]  # keep high TF-IDF words
    filtered_texts.append(tokens)

# --- 4. Rebuild Dictionary and Corpus After Filtering ---
id2word_filtered = corpora.Dictionary(filtered_texts)
corpus_final = [id2word_filtered.doc2bow(text) for text in filtered_texts]

# --- 5. Train LDA Model ---
num_topics = 20
lda_model = LdaModel(corpus=corpus_final, id2word=id2word_filtered, num_topics=num_topics, random_state=42, passes=10)

# --- 6. Visualize ---
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus_final, id2word_filtered)
vis
```

```python
#****************************************#
# TF-IDF + K-Means Clustering Pipeline
#****************************************#

#****************************************#
# 1. Import Libraries
#****************************************#
import re
import string
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import adjusted_rand_score
from nltk.corpus import stopwords

#****************************************#
# 2. Text Representation using TF-IDF
#****************************************#

# TF-IDF converts text into numerical vectors based on term frequency and inverse document frequency.
# This allows clustering algorithms to understand the "importance" of each word in a document relative to the corpus.

vectorizer = TfidfVectorizer(
    lowercase=True,            # convert text to lowercase
    stop_words="english",      # remove common English stopwords
    max_features=1000,          # limit to top 100 features for simplicity
    max_df=0.8,                # ignore terms that appear in more than 80% of documents
    min_df=2,                  # ignore terms that appear in less than 2 documents
    ngram_range=(1, 3)         # include unigrams, bigrams, and trigrams
)

# Fit the vectorizer on the 'file_content' column
vectors = vectorizer.fit_transform(df_copy['file_content'].fillna(''))
```

```python
#****************************************#
# 3. K-Means Clustering
#****************************************#

# K-Means groups similar documents into clusters based on TF-IDF features.
true_k = 20  # Number of clusters
k_model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=10, random_state=42)
k_model.fit(vectors)

#****************************************#
# 4. Extract Top Terms per Cluster
#****************************************#

# Identify the top words representing each cluster
order_centroids = k_model.cluster_centers_.argsort()[:, ::-1]
feature_names = vectorizer.get_feature_names_out()

top_features = {}
for i in range(true_k):
    top_indices = order_centroids[i, :10]   # top 10 words per cluster
    top_features[i] = [feature_names[ind] for ind in top_indices]
    print(f"Cluster {i}: {', '.join(top_features[i])}")

#****************************************#
# 5. Visualize Clusters in 2D Space
#****************************************#

# Reduce dimensions to 2D using PCA for visualization
pca = PCA(n_components=2, random_state=42)
reduced_features = pca.fit_transform(vectors.toarray())

# Create a DataFrame for plotting
df_plot = pd.DataFrame({
    'x': reduced_features[:, 0],
    'y': reduced_features[:, 1],
    'cluster': k_model.labels_
})

# Compute cluster centers and sizes for bubble plot
cluster_centers = df_plot.groupby('cluster')[['x','y']].mean().reset_index()
cluster_sizes = df_plot['cluster'].value_counts().sort_index().reset_index(drop=True)
cluster_centers['size'] = cluster_sizes
cluster_centers['keywords'] = cluster_centers['cluster'].apply(lambda x: ', '.join(top_features.get(x, [])))

# Plot using Plotly
fig = px.scatter(
    cluster_centers,
    x='x', y='y',
    size='size',
    color='cluster',
    hover_data={'keywords': True, 'size': True},
    text='cluster',
    title='K-Means Topic Clusters (TF-IDF)',
    size_max=80,
    color_continuous_scale='Viridis'
)
fig.update_traces(textposition='top center')
fig.update_layout(template='plotly_white', width=900, height=600)
fig.show()
```
