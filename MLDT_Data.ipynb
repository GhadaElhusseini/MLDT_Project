```python
# Text Representation using TF-IDF
vectorizer = TfidfVectorizer(lowercase=True, # no need to do it beforehand
                             stop_words = "english",
                             max_features = 100,
                             max_df = 0.8, # ignore terms that appear in more than 80% of the documents or if integer no float then "ignore terms that appear in more than 25 documents"
                             min_df = 2, # ignore terms that appear in less than # documents, min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
                             ngram_range=(1, 3) #remember bigrams and trigrams, this creates a tuple for this
                             )
vectors = vectorizer.fit_transform(df_clean['summary_database_enhancement_anonym_enhancement'].fillna(''))
```

```python
# We actually want to take the tf-idf and leverage it for topic modelling --> do it thru clustering
# k means to see any overlaps
```

```python
true_k = 20 # number of clusters
k_model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=10) # this the model we just then fit the model
k_model.fit(vectors) # pass in the vectors from the vectorizer
```

```python
order_centroids = k_model.cluster_centers_.argsort()[:, ::-1]
feature_names = vectorizer.get_feature_names_out()
```

```python
feature_names
```

```python
# Extract top terms for each cluster - this is just for observing the top features in each cluster
top_features = {}
for i in range(true_k):
    top_indices = order_centroids[i, :10]   # top 10 words per cluster
    top_features[i] = [feature_names[ind] for ind in top_indices]
    print(f"Cluster {i}: {', '.join(top_features[i])}")
```

```python
# Data Visualizations - using matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# Reduce dimensionality for plotting
pca = PCA(n_components=2, random_state=42)
reduced_features = pca.fit_transform(vectors.toarray())

# Create a dataframe with cluster labels
df_plot = pd.DataFrame({
    'x': reduced_features[:, 0],
    'y': reduced_features[:, 1],
    'cluster': k_model.labels_
})
```

```python
import plotly.express as px
import numpy as np

# Get cluster centers for bubbles
cluster_centers = df_plot.groupby('cluster')[['x','y']].mean().reset_index()
cluster_sizes = df_plot['cluster'].value_counts().sort_index().reset_index(drop=True)
cluster_centers['size'] = cluster_sizes
cluster_centers['keywords'] = cluster_centers['cluster'].apply(lambda x: ', '.join(top_features.get(x, [])))

fig = px.scatter(
    cluster_centers,
    x='x', y='y',
    size='size',
    color='cluster',
    hover_data={'keywords': True, 'size': True},
    text='cluster',
    title='K-Means Topic Clusters (TF-IDF)',
    size_max=80,
    color_continuous_scale='Viridis'
)
fig.update_traces(textposition='top center')
fig.update_layout(template='plotly_white', width=900, height=600)
fig.show()
```
